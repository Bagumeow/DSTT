```python
import numpy as np
import sympy
from sympy import Matrix

sympy.init_printing()
```

## Thu·∫≠t to√°n $ùê¥ = ùëÑùëÖ$

D√πng h√†m `np.linalg.qr()` ƒë·ªÉ ph√¢n r√£ $QR$ 1 ma tr·∫≠n

V√≠ d·ª•: D√πng thu·∫≠t gi·∫£i Gram-Schmidt ph√¢n r√£ $QR$ ma tr·∫≠n sau:
$$
\begin{bmatrix}
1 & -1 & 2\\
1 & 0 & -1\\
-1 & 1 & 2\\
0 & 1 & 1\\
\end{bmatrix}
$$

```python
A = Matrix([[1, -1, 2],
             [1, 0, -1],
             [-1, 1, 2],
             [0, 1, 1]])

Q, R = np.linalg.qr(A)
Matrix(Q), Matrix(R) # Ma tr·∫≠n Q v√† R
```

Ki·ªÉm tra l·∫°i thu·∫≠t to√°n $A = QR$ (c√≥ sai s·ªë)

```python
Matrix(Q) * Matrix(R) # Ma tr·∫≠n QR
```

D√πng thu·∫≠t gi·∫£i Gram-Schmidt ph√¢n r√£ $QR$ ma tr·∫≠n sau:
$$
\begin{bmatrix}
-2 & 1 & 3\\
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1\\
\end{bmatrix}
$$

```python
A = np.array([[-2, 1, 3],
             [1, 0, 0],
             [0, 1, 0],
             [0, 0, 1]])

Q, R = np.linalg.qr(A)
Matrix(Q), Matrix(R) # Ma tr·∫≠n Q v√† R
```

Ki·ªÉm tra l·∫°i thu·∫≠t to√°n $A = QR$ (c√≥ sai s·ªë)

```python
Matrix(Q) * Matrix(R) # Ma tr·∫≠n QR
```

## Thu·∫≠t to√°n $A = PDP^{-1}$

D√πng h√†m `sympy.Matrix.diagonalize()` ƒë·ªÉ ch√©o ho√° 1 ma tr·∫≠n. H√†m n√†y tr·∫£ v·ªÅ 2 ma tr·∫≠n ùëÉ v√† ùê∑ t∆∞∆°ng ·ª©ng

T√¨m $A^{20}$ th√¥ng qua thu·∫≠t to√°n $A = PDP^{-1}$:
$$A=
\begin{bmatrix}
-1 & -2 & 2\\
4 & 3 & -4\\
0 & -2 & 1\\
\end{bmatrix}
$$

```python
A = Matrix([[-1, -2, 2],
             [4, 3, -4],
             [0, -2, 1]])

P, D = A.diagonalize()
P * (D ** 20) * (P ** -1), A ** 20 # T√≠nh ma tr·∫≠n A^20 b·∫±ng 2 c√°ch
```

T√¨m $A^{10}$ th√¥ng qua thu·∫≠t to√°n $A = PDP^{-1}$:
$$A=
\begin{bmatrix}
1 & -2 & 0 & 5\\
0 & 7 & 1 & 5\\
0 & 4 & 4 & 0\\
0 & 0 & 0 & 2\\
\end{bmatrix}
$$

```python
A = Matrix([[1, -2, 0, 5],
             [0, 7, 1, 5],
             [0, 4, 4, 0],
            [0, 0, 0, 2]])

P, D = A.diagonalize()
P * (D ** 10) * (P ** -1), A ** 10 # T√≠nh ma tr·∫≠n A^10 b·∫±ng 2 c√°ch
```
